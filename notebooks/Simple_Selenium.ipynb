```python
import google_search as gs
gs.virtual_display_start()
import sst.actions as ssta
from selenium.webdriver.common.keys import Keys
```

```python
ssta.go_to('http://news.google.com/')
ssta.assert_title_contains('Google News')
```

```python
find_box = ssta.get_element(tag='input', id='gbqfq')
search_string = 'greek bonds'
find_box.send_keys(search_string + Keys.ENTER)
ssta.wait_for(ssta.assert_title_contains, search_string)
```

```python
search_results = ssta.get_elements(tag='td', css_class='tsw')
for r in search_results:
    heading = r.find_element_by_xpath('h3/a')
    print heading.text
```
